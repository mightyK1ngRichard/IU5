# Экспериментальное решение

````r
total_requests <- 1000000
arrival_times <- vector("numeric", total_requests) # времени прихода заявок
service_start_times <- vector("numeric", total_requests) # времени начала их обработки
queue_entry_times <- vector("numeric", total_requests) # времени входа и выхода из очереди,
queue_exit_times <- vector("numeric", total_requests) # времени завершения обработки,
completion_times <- vector("numeric", total_requests) # времени, когда сервер взял заявку и закончил её обслуживание
server_taken_times <- vector("numeric", total_requests)
server_done_times <- vector("numeric", total_requests)

# Генерируется время первого отказа канала и время его восстановления после отказа.
failure_time <- rexp(1, nu)
repair_time <- failure_time + rexp(1, gamma)

# Переменные состояния
queue_length <- 0 # текущая длина очереди.
queue_count <- 0 # всего поступило в очередь.
canceled_count <- 0 # потеряно (очередь была переполнена).
taken_count <- 0 #  сколько взято в обработку.
done_count <- 0 #  сколько успешно обработано.
queue_length_sum <- 0
free_count <- 0  # сколько заявок попало в свободный сервер.
empty_queue_count <- 0 # сколько раз заявка пришла, а очереди не было.
error_time <- 0 # суммарное время простоя канала из-за отказов.

# Генерация времен прихода заявок. Моделируется поток заявок — это пуассоновский процесс с интенсивностью λ.
arrival_times[1] <- rexp(1, lambda)
for (i in 2:total_requests) {
  arrival_times[i] <- arrival_times[i - 1] + rexp(1, lambda)
}

# Основной цикл обработки каждой заявки
for (t in arrival_times) {
  # Обработка, если сервер занят Заявка приходит, а сервер всё ещё обрабатывает предыдущую. Проверяется:
  # - Не наступил ли отказ во время текущего обслуживания.
  # - Если произошёл отказ и очередь полная, заявка теряется (canceled_count).
  # Иначе, ставится в очередь.
  if (taken_count > done_count) {
    start_time <- server_taken_times[take n_count]
    end_time <- server_done_times[taken_count]

    # Проверяется, был ли отказ сервера в момент обработки
    if (t >= failure_time && end_time >= failure_time) {
      done_count <- done_count + 1
      completion_times[done_count] <- failure_time
      server_done_times[done_count] <- repair_time

      # Если да и очередь полна — заявка теряется
      if (queue_length == m) {
        canceled_count <- canceled_count + 1 # очередь полна - заявка теряется
      } else {
        # Если нет — заявка добавляется в очередь.
        queue_count <- queue_count + 1
        queue_entry_times[queue_count] <- failure_time
        queue_length <- queue_length + 1
      }
    } else if (t > end_time) {
      done_count <- done_count + 1
      completion_times[done_count] <- end_time
    }
  }

  # Обработка очереди, если сервер освободился. Т.е пока очередь не пуста и сервер свободен
  while (done_count == taken_count && queue_length > 0) {
    # Пока есть заявки в очереди и сервер свободен:
    # - Извлекается первая заявка.
    if (done_count == 0) {
      #  Если это первая заявка вообще, стартуем с repair_time
      start_time <- repair_time
    } else {
      # Иначе — сервер освободится в server_done_times[done_count]
      start_time <- server_done_times[done_count]
    }
    # Проверка, не ушло ли время вперёд:
    if (start_time >= t) {
      break
    }

    # Определяется индекс в очереди и время начала обслуживания — не раньше, чем заявка пришла, и не раньше, чем сервер освободился.
    queue_position <- queue_count - queue_length + 1
    start_time <- max(queue_entry_times[queue_position], start_time)

    # устанавливаем время следуюющего отказа и восстановления. Обновляется failure_time, repair_time, если прошёл предыдущий отказ.
    while (start_time >= repair_time) {
      # Т.е если текущий start_time уже прошёл после восстановления сервера, генерим новые времена отказа и починки.
      failure_time <- repair_time + rexp(1, nu)
      repair_time <- failure_time + rexp(1, gamma)
    }

    # Если обслуживание начинается в момент, когда сервер сломался — учитываем время простоя (error_time) и откладываем обслуживание до восстановления.
    if (start_time >= failure_time && start_time <= repair_time) {
      error_time <- error_time + repair_time - failure_time
      if (repair_time < t)
        start_time <- repair_time
      else
        break
    }

    # Удаляем заявку из очереди, увеличиваем счётчик взятых, сохраняем времена начала и конца обслуживания.
    queue_length <- queue_length - 1
    taken_count <- taken_count + 1
    service_start_times[taken_count] <- queue_entry_times[queue_position]
    queue_exit_times[taken_count] <- start_time
    server_taken_times[taken_count] <- start_time
    end_time <- start_time + rexp(1, mu)
    server_done_times[taken_count] <- end_time

    # не сломался ли сервер во время нового обслуживания
    if (t >= failure_time && end_time >= failure_time) {
      done_count <- done_count + 1
      completion_times[done_count] <- failure_time
      server_done_times[done_count] <- repair_time
      #  Если отказ сервера случится в момент обслуживания:
      # - заявка всё равно считается как “завершённая” на момент отказа,
      # - сервер восстанавливается к repair_time.
      if (queue_length == m) {
        canceled_count <- canceled_count + 1
      } else {
        queue_count <- queue_count + 1
        queue_entry_times[queue_count] <- failure_time
        queue_length <- queue_length + 1
      }
    } else if (t > end_time) {
      done_count <- done_count + 1
      completion_times[done_count] <- end_time
    }
  }

  queue_length_sum <- queue_length_sum + queue_length

  # Если очередь пуста — увеличиваем счётчик пустых состояний очереди.
  if (queue_length == 0) {
    empty_queue_count <- empty_queue_count + 1
  }

   # Проверяем: свободен ли сервер
  if (done_count == taken_count) {
    start_time <- t

    # Генерация нового времени отказа/починки, если текущее время t уже после предыдущего восстановления
    while (start_time >= repair_time) {
      failure_time <- repair_time + rexp(1, nu)
      repair_time <- failure_time + rexp(1, gamma)
    }

    # Если сервер сломан в момент t
    if (start_time >= failure_time && start_time <= repair_time) {
      # Если очередь переполнена → заявка теряется:
      if (queue_length == m) {
        canceled_count <- canceled_count + 1
      } else {
        # Иначе заявка встаёт в очередь:
        queue_count <- queue_count + 1
        queue_entry_times[queue_count] <- start_time
        queue_length <- queue_length + 1
      }
    } else {
      # Сервер не сломан → заявка обслуживается сразу
      free_count <- free_count + 1
      taken_count <- taken_count + 1
      service_start_times[taken_count] <- start_time
      queue_exit_times[taken_count] <- start_time
      server_taken_times[taken_count] <- start_time
      end_time <- start_time + rexp(1, mu)
      server_done_times[taken_count] <- end_time
    }

  } else {
    # Значит, заявка идёт в очередь, если есть место, иначе — теряется
    if (queue_length == m) {
      canceled_count <- canceled_count + 1
    } else {
      queue_count <- queue_count + 1
      queue_entry_times[queue_count] <- t
      queue_length <- queue_length + 1
    }
  }
}

cat("Вероятность простоя:", free_count / total_requests, "\n")
cat("Вероятность образования очереди:", 1 - empty_queue_count / total_requests, "\n")
cat("Абсолютная пропускная способность:", (total_requests - canceled_count) / completion_times[done_count], "\n")
cat("Средняя длина очереди:", queue_length_sum / total_requests, "\n")
total_queue_time <- 0
for (i in 1:done_count) {
  total_queue_time <- total_queue_time + queue_exit_times[i] - service_start_times[i]
}
average_queue_time <- (total_queue_time - error_time) / total_requests
cat("Среднее время нахождения в очереди:", average_queue_time, "\n")
total_system_time <- 0
for (i in 1:done_count) {
  total_system_time <- total_system_time + completion_times[i] - service_start_times[i]
}
average_requests_in_system <- total_system_time / completion_times[done_count]
cat("Среднее число заявок в системе:", average_requests_in_system, "\n")
total_time_in_system <- 0
for (i in 1:done_count) {
  total_time_in_system <- completion_times[i] - service_start_times[i] + total_time_in_system
}
average_time_in_system <- total_time_in_system / total_requests
cat("Среднее время нахождения заявок в системе:", average_time_in_system, "\n")
````
